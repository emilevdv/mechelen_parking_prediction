### Academische onderbouwing

Datakwaliteit is een kritische maar vaak onderschatte stap in het opbouwen van 
betrouwbare voorspellingsmodellen. Tanui et al. (2025) tonen aan dat outlier-handling 
via ROLS (Recursive OLS) de modelnauwkeurigheid met tot 12% verbetert ten opzichte 
van ruwe data. Nezhadettehad et al. (2025) benadrukken dat sensoruitval en 
data-artefacten in parkeerdata systemisch zijn en leiden tot geflatteerde 
prestatiecijfers als ze niet worden gedetecteerd.

Specifiek voor tijdreeksdata van parkeersensoren zijn drie types anomalieën 
gedocumenteerd (Gao et al., 2023; Wan et al., 2023):
1. **Frozen sensor values**: de sensor rapporteert herhaaldelijk exact dezelfde 
   bezettingswaarde over meerdere opeenvolgende uren → wijst op sensor-uitval.
2. **Physically impossible values**: `occupancy_rate > 1.0` of 
   `occupied_spaces > capacity` → wijst op registratiefout of capaciteitswijziging.
3. **Temporal gaps**: ontbrekende uurobservaties per parking → problematisch voor 
   lag-features en autocorrelatie-modellen (LSTM, SARIMA).

We documenteren elke anomalie in een kwaliteitsrapport zodat beslissingen 
transparant en reproduceerbaar zijn — een vereiste voor academische validiteit.

## Stap 3 — Tijdreeksvolledigheid: verwachte vs. werkelijke observaties

Voor een correcte tijdreeksanalyse en het berekenen van lag-features is het 
essentieel dat de tijdreeks per parking geen onverwachte gaten heeft. 

**Verwacht**: elke parking heeft voor elke dag × 24 uur = 24 observaties.
Totaal verwacht per parking = aantal actieve dagen × 24.

Gaten in de tijdreeks zijn problematisch voor:
- Autocorrelatie-gebaseerde modellen (SARIMA, LSTM)
- Lag-features (lag_1h, lag_24h, lag_168h)
- Rolling-gemiddelden

We berekenen de completeness ratio per parking per jaar.


```python
# ── Verwachte uren per parking per jaar ───────────────────────────────────────
# Bouw een complete tijdreeks: voor elke parking, elke dag, elk uur

completeness_rows = []

for parking_id in df_st["parking_id"].unique():
    df_p = df_st[df_st["parking_id"] == parking_id].copy()

    for year in df_p["year"].unique():
        df_y = df_p[df_p["year"] == year]

        # Verwacht: alle unieke (date, hour) combinaties in dat jaar
        actual_dates   = df_y["date_only"].nunique()
        actual_obs     = len(df_y)
        expected_obs   = actual_dates * 24
        completeness   = actual_obs / expected_obs if expected_obs > 0 else 0
        n_missing      = expected_obs - actual_obs

        completeness_rows.append({
            "parking_id"    : parking_id,
            "year"          : year,
            "n_days"        : actual_dates,
            "expected_obs"  : expected_obs,
            "actual_obs"    : actual_obs,
            "missing_obs"   : n_missing,
            "completeness"  : round(completeness, 4)
        })

df_completeness = pd.DataFrame(completeness_rows)

print("=== Tijdreeksvolledigheid per parking per jaar ===")
print(df_completeness.to_string(index=False))

# ── Log parkings met <95% completeness ────────────────────────────────────────
low_coverage = df_completeness[df_completeness["completeness"] < 0.95]
log(
    check_name  = "Tijdreeksvolledigheid (<95%)",
    dataset     = "shortterm",
    result      = f"{len(low_coverage)} parking-jaar combinaties onder 95%",
    n_affected  = low_coverage["missing_obs"].sum(),
    action      = "Gedocumenteerd; geen verwijdering — periodes worden geflagd"
)
```

## Stap 4 — Duplicatendetectie

Een duplicaat in de tijdreeks is gedefinieerd als een rij met dezelfde 
combinatie van `(parking_id, rounded_hour)`. Het schema toont 54.436 unieke 
`rounded_hour` waarden tegenover 697.218 rijen en 10 parkings → 
theoretisch maximum is 54.436 × 10 = 544.360 unieke combinaties. 
De 697.218 rijen suggereren mogelijke duplicaten of meerdere observaties 
per uurslot (bijv. correctie-records).


```python
# ── Exacte duplicaten (alle 18 kolommen identiek) ─────────────────────────────
n_exact_dupes = df_st.duplicated().sum()
log("Exacte duplicaten", "shortterm", f"{n_exact_dupes} rijen", 
    n_exact_dupes, "Verwijderen" if n_exact_dupes > 0 else "Geen actie")

# ── Functionele duplicaten: zelfde (parking_id, rounded_hour) ─────────────────
n_func_dupes = df_st.duplicated(subset=["parking_id","rounded_hour"]).sum()
log("Functionele duplicaten (parking_id + rounded_hour)", "shortterm",
    f"{n_func_dupes} rijen", n_func_dupes,
    "Laatste record bewaren (meest recente publicatie)" if n_func_dupes > 0 else "Geen actie")

if n_exact_dupes > 0:
    df_st = df_st.drop_duplicates()
    print(f"✓ Exacte duplicaten verwijderd. Nieuwe shape: {df_st.shape}")

if n_func_dupes > 0:
    # Bewaar het record met de meest recente publication_time
    df_st = (df_st
             .sort_values("publication_time", ascending=False)
             .drop_duplicates(subset=["parking_id","rounded_hour"])
             .sort_values(["parking_id","rounded_hour"])
             .reset_index(drop=True))
    print(f"✓ Functionele duplicaten verwijderd. Nieuwe shape: {df_st.shape}")
```

## Stap 5 — Bezettingswaarden: fysische plausibiliteit

Op basis van het schema is `occupancy_rate` reeds begrensd [0, 1] in de shortterm 
data. Toch controleren we:

1. **Interne consistentie**: is `occupancy_rate ≈ occupied / override`?  
   Tolerantie: |berekende rate - opgeslagen rate| > 0.01 wijst op inconsistentie.
2. **Negatieve waarden**: `number_of_occupied_spaces < 0` is fysisch onmogelijk.
3. **Bezetting > override**: `occupied > number_of_spaces_override`.
4. **Frozen sensor**: ≥ 4 opeenvolgende identieke bezettingswaarden per parking 
   (Gao et al., 2023 hanteren een vergelijkbaar criterium voor sensor-uitval).

Voor de longterm-data is de occupancy_rate-kolom deels NULL (zie schema) — 
dit vereist aparte afhandeling.


```python
# ── Check 1: Interne consistentie occupancy_rate ──────────────────────────────
df_st["occ_rate_calculated"] = (
    df_st["number_of_occupied_spaces"] / df_st["number_of_spaces_override"]
).replace([np.inf, -np.inf], np.nan)

df_st["occ_rate_discrepancy"] = (
    (df_st["occupancy_rate"] - df_st["occ_rate_calculated"]).abs()
)

n_inconsistent = (df_st["occ_rate_discrepancy"] > 0.01).sum()
log("Occupancy rate inconsistentie (>0.01)", "shortterm",
    f"{n_inconsistent} rijen ({n_inconsistent/len(df_st)*100:.2f}%)",
    n_inconsistent, "Flaggen als 'occ_rate_inconsistent'")

df_st["flag_occ_inconsistent"] = (df_st["occ_rate_discrepancy"] > 0.01).astype(int)

# ── Check 2: Bezetting > override ─────────────────────────────────────────────
n_overcapacity = (df_st["number_of_occupied_spaces"] > df_st["number_of_spaces_override"]).sum()
log("Bezetting > capaciteitsoverride", "shortterm",
    f"{n_overcapacity} rijen", n_overcapacity, "Flaggen als 'flag_overcapacity'")

df_st["flag_overcapacity"] = (
    df_st["number_of_occupied_spaces"] > df_st["number_of_spaces_override"]
).astype(int)

# ── Check 3: Negatieve bezetting ──────────────────────────────────────────────
n_negative = (df_st["number_of_occupied_spaces"] < 0).sum()
log("Negatieve bezetting", "shortterm", f"{n_negative} rijen", n_negative,
    "Verwijderen" if n_negative > 0 else "Geen actie")

# ── Opruimen tijdelijke kolom ─────────────────────────────────────────────────
df_st = df_st.drop(columns=["occ_rate_calculated","occ_rate_discrepancy"])

print(f"\nFlag-verdeling:\n{df_st[['flag_occ_inconsistent','flag_overcapacity']].sum()}")
```

```python
# ── Frozen sensor: ≥ 4 opeenvolgende identieke occupied-waarden per parking ───
FREEZE_THRESHOLD = 4

df_st = df_st.sort_values(["parking_id","rounded_hour"]).reset_index(drop=True)

df_st["prev_occupied"] = df_st.groupby("parking_id")["number_of_occupied_spaces"].shift(1)
df_st["is_same_as_prev"] = (
    df_st["number_of_occupied_spaces"] == df_st["prev_occupied"]
).astype(int)

# Cumulatieve teller die reset bij verandering
df_st["freeze_streak"] = (
    df_st.groupby(
        ["parking_id",
         (df_st["is_same_as_prev"] != df_st.groupby("parking_id")["is_same_as_prev"].shift(1, fill_value=0)).cumsum()]
    )["is_same_as_prev"].cumsum()
)

df_st["flag_frozen_sensor"] = (df_st["freeze_streak"] >= FREEZE_THRESHOLD).astype(int)

n_frozen = df_st["flag_frozen_sensor"].sum()
log("Frozen sensor (≥4 identieke opeenvolgende uren)", "shortterm",
    f"{n_frozen} rijen ({n_frozen/len(df_st)*100:.2f}%)",
    n_frozen, "Flaggen als 'flag_frozen_sensor'; niet verwijderd")

print(f"\nFrozen sensor rijen per parking:")
print(df_st.groupby("parking_id")["flag_frozen_sensor"].sum().sort_values(ascending=False))

# Opruimen hulpkolommen
df_st = df_st.drop(columns=["prev_occupied","is_same_as_prev","freeze_streak"])
```

## Stap 6 — Zomertijdcheck (DST)

België schakelt twee keer per jaar om:
- **Zomertijd**: laatste zondag van maart (klok +1u → uur 02:00 ontbreekt)
- **Wintertijd**: laatste zondag van oktober (klok -1u → uur 02:00 dubbelzinnig)

`publication_time` is TIMESTAMP_NS (timezone-naïef), `created_at` heeft wel 
timezone. Rond de DST-overgangen verwachten we anomalieën in de tijdreeks:
- Ontbrekend uur in maart → gap van 2 uur
- Dubbel uur in oktober → mogelijke duplicaat of sprong

We detecteren dit door per parking per dag te controleren of er 
exact 24 unieke uurwaarden aanwezig zijn.


```python
# DST-overgangsdatums 2019–2026 (laatste zondag maart/oktober)
DST_TRANSITIONS = [
    # (datum, type)
    ("2019-03-31","zomer→"), ("2019-10-27","←winter"),
    ("2020-03-29","zomer→"), ("2020-10-25","←winter"),
    ("2021-03-28","zomer→"), ("2021-10-31","←winter"),
    ("2022-03-27","zomer→"), ("2022-10-30","←winter"),
    ("2023-03-26","zomer→"), ("2023-10-29","←winter"),
    ("2024-03-31","zomer→"), ("2024-10-27","←winter"),
    ("2025-03-30","zomer→"), ("2025-10-26","←winter"),
    ("2026-03-29","zomer→"), ("2026-10-25","←winter"),
]

dst_dates = [pd.Timestamp(d[0]).date() for d in DST_TRANSITIONS]

# Uurtellingen per parking per dag
hours_per_day = (df_st
    .groupby(["parking_id","date_only"])["hour"]
    .nunique()
    .reset_index(name="n_unique_hours"))

# Filter op DST-datums
dst_check = hours_per_day[hours_per_day["date_only"].isin(dst_dates)]

print("=== Uurtellingen op DST-overgangsdagen ===")
if len(dst_check) > 0:
    print(dst_check.to_string(index=False))
else:
    print("Geen DST-datums aanwezig in data of alle uren compleet.")

n_dst_anom = (dst_check["n_unique_hours"] != 24).sum()
log("DST-uuranomalieën", "shortterm", f"{n_dst_anom} parking-dag combinaties",
    n_dst_anom, "Gedocumenteerd; geen verwijdering")
```

## Stap 7 — `timeseries_longterm.parquet` kwaliteitscheck

De longterm-data heeft een aantal specifieke kenmerken:
- **7 parkings** (subset van de 10 in shortterm — niet alle parkings hebben abonnees)
- **Enkel 2024** (year distinct count = 1)
- `number_of_spaces_override`: min=20, max=100 → veel kleiner dan shortterm 
  (abonneesplaatsen zijn een gereserveerde subset van de totale capaciteit)
- `occupancy_rate`: heeft NULL-waarden (zie schema)

Conceptueel onderscheid (belangrijk voor modellering):
- Shortterm: fluctuerende vraag, sterk afhankelijk van externe factoren 
  (weer, evenementen, vakantie) → hoofddataset voor RQ1
- Longterm: structurele abonneesbezetting, meer stabiel en inertiegedreven → 
  supplementaire analyse of aparte modelleringsdoelstelling


```python
df_lt = pd.read_parquet(DATA_RAW / "timeseries_longterm.parquet")

print(f"Shape         : {df_lt.shape}")
print(f"\nNull-tellingen:\n{df_lt.isnull().sum()}")
print(f"\nJaar(en)      : {df_lt['year'].unique()}")
print(f"\nParkings ({df_lt['parking_id'].nunique()}):")
print(df_lt["parking_id"].value_counts())

print(f"\nnumber_of_spaces_override statistieken:")
print(df_lt["number_of_spaces_override"].describe())

print(f"\nnumber_of_occupied_spaces statistieken:")
print(df_lt["number_of_occupied_spaces"].describe())
```

```python
# ── Occupancy rate NULL-waarden: herberekenen ─────────────────────────────────
n_null_occ = df_lt["occupancy_rate"].isnull().sum()
log("Null occupancy_rate", "longterm", f"{n_null_occ} nulls",
    n_null_occ, "Herberekenen als occupied/override")

df_lt["occupancy_rate"] = df_lt["occupancy_rate"].fillna(
    df_lt["number_of_occupied_spaces"] / df_lt["number_of_spaces_override"]
)

# Check na imputatie
n_still_null = df_lt["occupancy_rate"].isnull().sum()
print(f"Null na imputatie: {n_still_null}")

# ── Consistentiecheck ─────────────────────────────────────────────────────────
df_lt["occ_calc"] = df_lt["number_of_occupied_spaces"] / df_lt["number_of_spaces_override"]
n_incons_lt = ((df_lt["occupancy_rate"] - df_lt["occ_calc"]).abs() > 0.01).sum()
log("Occupancy inconsistentie", "longterm", f"{n_incons_lt} rijen",
    n_incons_lt, "Flaggen")

df_lt["flag_occ_inconsistent"] = (
    (df_lt["occupancy_rate"] - df_lt["occ_calc"]).abs() > 0.01
).astype(int)
df_lt = df_lt.drop(columns=["occ_calc"])

# ── Frozen sensor check longterm ───────────────────────────────────────────────
df_lt = df_lt.sort_values(["parking_id","rounded_hour"]).reset_index(drop=True)
df_lt["prev_occ"] = df_lt.groupby("parking_id")["number_of_occupied_spaces"].shift(1)
df_lt["is_same"]  = (df_lt["number_of_occupied_spaces"] == df_lt["prev_occ"]).astype(int)
df_lt["freeze_streak"] = (
    df_lt.groupby(
        ["parking_id",
         (df_lt["is_same"] != df_lt.groupby("parking_id")["is_same"].shift(1, fill_value=0)).cumsum()]
    )["is_same"].cumsum()
)
df_lt["flag_frozen_sensor"] = (df_lt["freeze_streak"] >= FREEZE_THRESHOLD).astype(int)

n_frozen_lt = df_lt["flag_frozen_sensor"].sum()
log("Frozen sensor longterm", "longterm", f"{n_frozen_lt} rijen",
    n_frozen_lt, "Flaggen")

df_lt = df_lt.drop(columns=["prev_occ","is_same","freeze_streak"])

# ── Duplicaten ────────────────────────────────────────────────────────────────
n_dupes_lt = df_lt.duplicated(subset=["parking_id","rounded_hour"]).sum()
log("Functionele duplicaten", "longterm", f"{n_dupes_lt}", n_dupes_lt,
    "Verwijderen" if n_dupes_lt > 0 else "Geen actie")

if n_dupes_lt > 0:
    df_lt = (df_lt.sort_values("publication_time", ascending=False)
                  .drop_duplicates(subset=["parking_id","rounded_hour"])
                  .sort_values(["parking_id","rounded_hour"])
                  .reset_index(drop=True))
```

## Stap 8 — Capaciteitsoverride vs. locatiedata kruischeck

`number_of_spaces_override` in de tijdreeks is de **operationele** capaciteit 
op dat uurmoment. `total_capacity` in `parking_location` is de **totale** 
gecertificeerde capaciteit.

Relatie: `override ≤ total_capacity` altijd.  
Uitzondering: tijdelijke capaciteitsuitbreidingen of data-invoerfouten.

We kruisen beide om parkings te identificeren waar de override-waarde de 
gecertificeerde capaciteit overstijgt.


```python
# Maximale override per parking in shortterm
max_override_st = (df_st.groupby("parking_id")["number_of_spaces_override"]
                        .max()
                        .reset_index(name="max_override"))

# Join met locatiedata
cap_check = max_override_st.merge(
    df_loc[["parking_id","total_capacity"]],
    on="parking_id", how="left"
)
cap_check["override_exceeds_capacity"] = (
    cap_check["max_override"] > cap_check["total_capacity"]
)

print("=== Capaciteitskruischeck (shortterm max override vs. total_capacity) ===")
print(cap_check.to_string(index=False))

n_exceeds = cap_check["override_exceeds_capacity"].sum()
log("Override > total_capacity", "shortterm",
    f"{n_exceeds} parkings", n_exceeds,
    "Documenteren; mogelijke dynamische capaciteitswijziging of datafout")
```

## Stap 9 — Overzicht van alle kwaliteitsflags

De gecleande datasets bevatten de volgende boolean-flags (0/1):

| Flag | Betekenis |
|---|---|
| `flag_occ_inconsistent` | `\|occupancy_rate - occupied/override\| > 0.01` |
| `flag_overcapacity` | `occupied > override` (shortterm only) |
| `flag_frozen_sensor` | ≥ 4 opeenvolgende identieke bezettingswaarden |

**Beleid voor modellering**: rijen met `flag_frozen_sensor = 1` worden 
**uitgesloten** van modeltraining (niet verwijderd uit dataset). 
Rijen met `flag_occ_inconsistent = 1` worden bewaard maar de 
`occupancy_rate` wordt herberekend vanuit `occupied/override`.


```python
# Herbereken occupancy_rate op basis van de ruwe telwaarden (betrouwbaarder)
for df, name in [(df_st, "shortterm"), (df_lt, "longterm")]:
    mask = df["flag_occ_inconsistent"] == 1
    n = mask.sum()
    if n > 0:
        df.loc[mask, "occupancy_rate"] = (
            df.loc[mask, "number_of_occupied_spaces"] /
            df.loc[mask, "number_of_spaces_override"]
        )
        print(f"[{name}] {n} occupancy_rate waarden gecorrigeerd via herberekening.")
    else:
        print(f"[{name}] Geen correctie nodig.")
```

```python
print("=== FLAG SAMENVATTING SHORTTERM ===")
flag_cols_st = [c for c in df_st.columns if c.startswith("flag_")]
print(df_st[flag_cols_st].sum().to_frame("n_flagged"))
print(f"\nTotaal bruikbare rijen (geen frozen sensor): "
      f"{(df_st['flag_frozen_sensor'] == 0).sum():,}")

print("\n=== FLAG SAMENVATTING LONGTERM ===")
flag_cols_lt = [c for c in df_lt.columns if c.startswith("flag_")]
print(df_lt[flag_cols_lt].sum().to_frame("n_flagged"))
print(f"\nTotaal bruikbare rijen (geen frozen sensor): "
      f"{(df_lt['flag_frozen_sensor'] == 0).sum():,}")
```
