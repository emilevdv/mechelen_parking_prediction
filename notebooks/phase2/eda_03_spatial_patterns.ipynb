```python
from __future__ import annotations

from pathlib import Path
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.spatial.distance import squareform
from scipy.stats import spearmanr, theilslopes

pd.set_option("display.max_columns", 200)
pd.set_option("display.max_rows", 200)
sns.set_theme(style="whitegrid", context="notebook", palette="colorblind")


def find_project_root(start: Path | None = None) -> Path:
    current = (start or Path.cwd()).resolve()
    for candidate in (current, *current.parents):
        if (candidate / "data_processed").exists() and (candidate / "notebooks").exists():
            return candidate
    raise FileNotFoundError("Project root not found from current working directory")


def as_flag(series: pd.Series) -> pd.Series:
    return pd.to_numeric(series, errors="coerce").fillna(0).gt(0)


def build_quality_mask(df: pd.DataFrame) -> pd.Series:
    mask = pd.Series(True, index=df.index)
    for col in ["system_blackout", "low_data_coverage", "partial_year", "flag_occ_inconsistent"]:
        if col in df.columns:
            mask &= ~as_flag(df[col])
    mask &= pd.to_numeric(df["occupancy_rate"], errors="coerce").between(0, 1, inclusive="both")
    return mask


def haversine_km(lat1, lon1, lat2, lon2):
    r = 6371.0
    p = np.pi / 180.0
    dlat = (lat2 - lat1) * p
    dlon = (lon2 - lon1) * p
    a = np.sin(dlat / 2.0) ** 2 + np.cos(lat1 * p) * np.cos(lat2 * p) * np.sin(dlon / 2.0) ** 2
    return 2 * r * np.arcsin(np.sqrt(a))


PROJECT_ROOT = find_project_root()

# === AUTO-EXPORT ARTIFACTS (figures + displayed tables) ===
NOTEBOOK_SLUG = "eda_03_spatial_patterns"
ARTIFACTS_DIR = PROJECT_ROOT / "artifacts" / "phase2" / NOTEBOOK_SLUG
FIG_DIR = ARTIFACTS_DIR / "figures"
TABLE_DIR = ARTIFACTS_DIR / "tables"
LOG_DIR = ARTIFACTS_DIR / "logs"

for _d in [ARTIFACTS_DIR, FIG_DIR, TABLE_DIR, LOG_DIR]:
    _d.mkdir(parents=True, exist_ok=True)


def _safe_artifact_name(name: str) -> str:
    allowed = set("abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789_-")
    s = "".join(ch if ch in allowed else "_" for ch in str(name))
    while "__" in s:
        s = s.replace("__", "_")
    s = s.strip("_")
    return s or "artifact"


def save_dataframe_artifact(df: pd.DataFrame, name: str, index: bool = True) -> dict[str, str | None]:
    base = _safe_artifact_name(name)
    csv_path = TABLE_DIR / f"{base}.csv"
    parquet_path = TABLE_DIR / f"{base}.parquet"

    df.to_csv(csv_path, index=index)
    parquet_ok = True
    try:
        df.to_parquet(parquet_path, index=index)
    except Exception:
        parquet_ok = False

    return {
        "csv": str(csv_path),
        "parquet": str(parquet_path) if parquet_ok else None,
    }


if not globals().get("_DISPLAY_AUTO_EXPORT_PATCHED", False):
    _DISPLAY_AUTO_EXPORT_PATCHED = True
    _ORIG_DISPLAY = display
    _DISPLAY_COUNTER = {"n": 0}

    def display(*objs, **kwargs):
        for obj in objs:
            if isinstance(obj, pd.DataFrame):
                _DISPLAY_COUNTER["n"] += 1
                save_dataframe_artifact(obj, f"display_{_DISPLAY_COUNTER['n']:03d}", index=True)
        return _ORIG_DISPLAY(*objs, **kwargs)


try:
    import matplotlib.pyplot as plt  # noqa: F401

    if not getattr(plt, "_AUTO_EXPORT_PATCHED", False):
        _ORIG_PLT_SHOW = plt.show
        _FIG_COUNTER = {"n": 0}

        def _show_and_save(*args, **kwargs):
            fig_nums = list(plt.get_fignums())
            for fig_num in fig_nums:
                fig = plt.figure(fig_num)
                _FIG_COUNTER["n"] += 1
                fig_path = FIG_DIR / f"fig_{_FIG_COUNTER['n']:03d}.png"
                fig.savefig(fig_path, dpi=150, bbox_inches="tight")
            return _ORIG_PLT_SHOW(*args, **kwargs)

        plt.show = _show_and_save
        plt._AUTO_EXPORT_PATCHED = True
    FIG_EXPORT_ENABLED = True
except Exception:
    FIG_EXPORT_ENABLED = False

print(f"Artifacts directory: {ARTIFACTS_DIR}")
print(f"- Figures: {FIG_DIR}")
print(f"- Tables: {TABLE_DIR}")

DATA_DIR = PROJECT_ROOT / "data_processed"

DATA_PATHS = {
    "shortterm": DATA_DIR / "MAD_shortterm.parquet",
    "longterm": DATA_DIR / "MAD_longterm.parquet",
}

raw_dfs = {}
filtered_dfs = {}

for dataset_label, path in DATA_PATHS.items():
    if not path.exists():
        raise FileNotFoundError(f"Dataset ontbreekt: {dataset_label} -> {path}")

    df = pd.read_parquet(path).copy()
    df["dataset_label"] = dataset_label
    df["rounded_hour"] = pd.to_datetime(df["rounded_hour"], errors="coerce")
    df["date_only"] = pd.to_datetime(df["date_only"], errors="coerce")

    df["tier_temporal"] = np.where(
        df["parking_location_category"].astype(str).eq("centrum"),
        "centrum",
        "vesten_of_rand",
    )

    raw_dfs[dataset_label] = df
    filtered_dfs[dataset_label] = df.loc[build_quality_mask(df)].copy()

overview_rows = []
for label in ["shortterm", "longterm"]:
    d0 = raw_dfs[label]
    d1 = filtered_dfs[label]
    overview_rows.append(
        {
            "dataset_label": label,
            "n_raw": int(len(d0)),
            "n_quality_filtered": int(len(d1)),
            "pct_retained": float(len(d1) / len(d0) * 100),
            "n_parkings": int(d1["parking_id"].nunique()),
            "date_min": d1["rounded_hour"].min(),
            "date_max": d1["rounded_hour"].max(),
            "years": ", ".join(map(str, sorted(d1["year"].dropna().unique().tolist()))),
        }
    )

overview_df = pd.DataFrame(overview_rows).set_index("dataset_label")
display(overview_df)
```

## 2. Vergelijking tussen tiers en location categories

```python
def tier_location_summary(df: pd.DataFrame, by_col: str) -> pd.DataFrame:
    out = (
        df.groupby(["dataset_label", by_col])["occupancy_rate"]
        .agg(
            n_obs="size",
            mean="mean",
            median="median",
            std="std",
            iqr=lambda s: s.quantile(0.75) - s.quantile(0.25),
            q95=lambda s: s.quantile(0.95),
            q99=lambda s: s.quantile(0.99),
        )
        .reset_index()
    )
    out["cv"] = out["std"] / out["mean"]
    return out


tier_summary = tier_location_summary(all_filtered, "tier_temporal")
location_summary = tier_location_summary(all_filtered, "parking_location_category")

print("Tier summary")
display(tier_summary.round(4))

print("Location summary")
display(location_summary.round(4))

# Effectgrootte centrum vs vesten_of_rand
center_outer_diff_rows = []
for dataset_label, grp in all_filtered.groupby("dataset_label"):
    center = grp.loc[grp["tier_temporal"] == "centrum", "occupancy_rate"]
    outer = grp.loc[grp["tier_temporal"] == "vesten_of_rand", "occupancy_rate"]
    center_outer_diff_rows.append(
        {
            "dataset_label": dataset_label,
            "mean_diff_center_minus_outer": float(center.mean() - outer.mean()),
            "median_diff_center_minus_outer": float(center.median() - outer.median()),
            "q95_diff_center_minus_outer": float(center.quantile(0.95) - outer.quantile(0.95)),
            "std_diff_center_minus_outer": float(center.std() - outer.std()),
        }
    )

center_outer_diff = pd.DataFrame(center_outer_diff_rows).set_index("dataset_label")
print("Centrum minus vesten_of_rand (samenvatting)")
display(center_outer_diff.round(4))
```

## 3. Relatie tussen capaciteit en occupancy_rate

```python
capacity_relation_rows = []
parking_level_rows = []

for dataset_label, df in filtered_dfs.items():
    rho_row, p_row = spearmanr(df["total_capacity"], df["occupancy_rate"])

    parking_level = (
        df.groupby("parking_id")
        .agg(
            total_capacity=("total_capacity", "median"),
            mean_occupancy=("occupancy_rate", "mean"),
            q95_occupancy=("occupancy_rate", lambda s: s.quantile(0.95)),
            std_occupancy=("occupancy_rate", "std"),
            tier_temporal=("tier_temporal", "first"),
            parking_location_category=("parking_location_category", "first"),
        )
        .reset_index()
    )

    rho_parking, p_parking = spearmanr(parking_level["total_capacity"], parking_level["mean_occupancy"])

    slope, intercept, slope_low, slope_high = theilslopes(
        parking_level["mean_occupancy"],
        parking_level["total_capacity"],
    )

    capacity_relation_rows.append(
        {
            "dataset_label": dataset_label,
            "row_level_spearman_rho": float(rho_row),
            "row_level_spearman_p": float(p_row),
            "parking_level_spearman_rho": float(rho_parking),
            "parking_level_spearman_p": float(p_parking),
            "theil_sen_slope_parking_level": float(slope),
            "theil_sen_slope_low": float(slope_low),
            "theil_sen_slope_high": float(slope_high),
        }
    )

    parking_level["dataset_label"] = dataset_label
    parking_level_rows.append(parking_level)

capacity_relation = pd.DataFrame(capacity_relation_rows).set_index("dataset_label")
parking_level_capacity = pd.concat(parking_level_rows, ignore_index=True)

print("Capaciteit-occupancy relatie")
display(capacity_relation.round(6))

print("Parking-level tabel voor capaciteitrelatie")
display(parking_level_capacity.sort_values(["dataset_label", "total_capacity"]).round(4))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

for ax, dataset_label in zip(axes, ["shortterm", "longterm"]):
    sub = parking_level_capacity.loc[parking_level_capacity["dataset_label"] == dataset_label].copy()

    sns.scatterplot(
        data=sub,
        x="total_capacity",
        y="mean_occupancy",
        hue="tier_temporal",
        style="parking_location_category",
        s=150,
        ax=ax,
    )

    # Robuuste Theil-Sen lijn
    slope = capacity_relation.loc[dataset_label, "theil_sen_slope_parking_level"]
    x = np.linspace(sub["total_capacity"].min(), sub["total_capacity"].max(), 100)
    intercept = np.median(sub["mean_occupancy"] - slope * sub["total_capacity"])
    y = intercept + slope * x
    ax.plot(x, y, color="black", linestyle="--", label="Theil-Sen")

    for _, row in sub.iterrows():
        ax.text(row["total_capacity"], row["mean_occupancy"], row["parking_id"], fontsize=8)

    ax.set_title(f"Capaciteit vs gemiddelde occupancy ({dataset_label})")
    ax.set_xlabel("total_capacity")
    ax.set_ylabel("mean_occupancy")
    ax.legend(loc="best")

plt.tight_layout()
plt.show()
```

**Interpretatie**

H-S2 krijgt geen uniforme steun: in longterm is de negatieve capaciteitsrelatie duidelijker, terwijl shortterm op parkingniveau een zwakke en onzekere relatie toont. Dit wijst erop dat capaciteit op zichzelf onvoldoende is als generieke spatial predictor zonder interactie met parkingcontext.

**Hypothese status**
- H-S2: **Partially supported** (sterker in longterm, zwak/inconclusief in shortterm).

**Volgende logische stap**
- Similariteit tussen parkings onderzoeken om te beoordelen of tiersegmentatie voldoende coherent is (H-S3).

## 4. Similariteit en correlatie tussen parkings

```python
def make_profile_corr(df: pd.DataFrame) -> tuple[pd.DataFrame, pd.DataFrame]:
    profile = (
        df.groupby(["parking_id", "weekday_int", "hour"])["occupancy_rate"]
        .mean()
        .reset_index()
    )
    profile["slot"] = profile["weekday_int"].astype(int) * 24 + profile["hour"].astype(int)

    mat = profile.pivot(index="parking_id", columns="slot", values="occupancy_rate").sort_index()
    mat_z = mat.sub(mat.mean(axis=1), axis=0).div(mat.std(axis=1).replace(0, np.nan), axis=0)
    corr = mat_z.T.corr()
    return mat_z, corr


def pairwise_tier_corr_stats(corr: pd.DataFrame, tier_map: pd.Series) -> pd.DataFrame:
    rows = []
    ids = corr.index.tolist()
    for i, a in enumerate(ids):
        for b in ids[i + 1 :]:
            rows.append(
                {
                    "pair": f"{a}__{b}",
                    "corr": float(corr.loc[a, b]),
                    "same_tier": bool(tier_map.loc[a] == tier_map.loc[b]),
                }
            )
    out = pd.DataFrame(rows)
    summary = (
        out.groupby("same_tier")["corr"]
        .agg(["count", "mean", "median", "std"])
        .rename(index={True: "within_tier", False: "between_tier"})
    )
    return out, summary


profile_mats = {}
profile_corrs = {}
profile_pair_summaries = {}
profile_pair_detail = {}

for dataset_label, df in filtered_dfs.items():
    mat_z, corr = make_profile_corr(df)
    profile_mats[dataset_label] = mat_z
    profile_corrs[dataset_label] = corr

    tiers = df.groupby("parking_id")["tier_temporal"].first()
    detail, summary = pairwise_tier_corr_stats(corr, tiers)

    detail["dataset_label"] = dataset_label
    summary = summary.reset_index().rename(columns={"same_tier": "pair_type"})
    summary["dataset_label"] = dataset_label

    profile_pair_detail[dataset_label] = detail
    profile_pair_summaries[dataset_label] = summary

print("Correlatiematrices van gestandaardiseerde weekprofielen")
for dataset_label in ["shortterm", "longterm"]:
    print(f"\n--- {dataset_label} ---")
    display(profile_corrs[dataset_label].round(3))

pair_summary_df = pd.concat(profile_pair_summaries.values(), ignore_index=True)
print("Binnen-tier versus tussen-tier profielcorrelaties")
display(pair_summary_df.round(4))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

for ax, dataset_label in zip(axes, ["shortterm", "longterm"]):
    corr = profile_corrs[dataset_label]
    sns.heatmap(corr, vmin=-1, vmax=1, cmap="coolwarm", center=0, annot=True, fmt=".2f", ax=ax)
    ax.set_title(f"Weekprofiel correlatiematrix ({dataset_label})")

plt.tight_layout()
plt.show()

# Dendrogram op basis van (1 - correlatie)
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

for ax, dataset_label in zip(axes, ["shortterm", "longterm"]):
    corr = profile_corrs[dataset_label]
    dist_df = (1 - corr).fillna(1.0).clip(lower=0.0)
    dist_arr = dist_df.to_numpy(copy=True)
    np.fill_diagonal(dist_arr, 0.0)

    condensed = squareform(dist_arr, checks=False)
    Z = linkage(condensed, method="average")
    dendrogram(Z, labels=corr.index.tolist(), leaf_rotation=60, ax=ax)
    ax.set_title(f"Clustering van parkings ({dataset_label})")
    ax.set_ylabel("distance = 1 - corr")

plt.tight_layout()
plt.show()
```

```python
# Alternatieve similarity-check op echte overlappende uurreeksen + afstandsrelatie (H-S7)
distance_similarity_rows = []

ts_pair_summary_rows = []
for dataset_label, df in filtered_dfs.items():
    wide = df.pivot_table(index="rounded_hour", columns="parking_id", values="occupancy_rate", aggfunc="mean")
    z = (wide - wide.mean()) / wide.std()
    corr_ts = z.corr(min_periods=200)

    tier_map = df.groupby("parking_id")["tier_temporal"].first()
    detail_ts, summary_ts = pairwise_tier_corr_stats(corr_ts, tier_map)
    summary_ts = summary_ts.reset_index().rename(columns={"same_tier": "pair_type"})
    summary_ts["dataset_label"] = dataset_label
    ts_pair_summary_rows.append(summary_ts)

    coords = (
        df.groupby("parking_id")
        .agg(latitude=("latitude", "median"), longitude=("longitude", "median"))
        .reindex(corr_ts.index)
    )

    ids = corr_ts.index.tolist()
    for i, a in enumerate(ids):
        for b in ids[i + 1 :]:
            dist_km = haversine_km(
                coords.loc[a, "latitude"],
                coords.loc[a, "longitude"],
                coords.loc[b, "latitude"],
                coords.loc[b, "longitude"],
            )
            distance_similarity_rows.append(
                {
                    "dataset_label": dataset_label,
                    "pair": f"{a}__{b}",
                    "dist_km": float(dist_km),
                    "corr_ts": float(corr_ts.loc[a, b]),
                    "same_tier": bool(tier_map.loc[a] == tier_map.loc[b]),
                }
            )


ts_pair_summary_df = pd.concat(ts_pair_summary_rows, ignore_index=True)
distance_similarity_df = pd.DataFrame(distance_similarity_rows)

distance_corr_rows = []
for dataset_label, grp in distance_similarity_df.groupby("dataset_label"):
    rho, p = spearmanr(grp["dist_km"], grp["corr_ts"])
    distance_corr_rows.append(
        {
            "dataset_label": dataset_label,
            "spearman_rho_distance_vs_corr": float(rho),
            "spearman_p": float(p),
            "n_pairs": int(len(grp)),
        }
    )

distance_corr_summary = pd.DataFrame(distance_corr_rows).set_index("dataset_label")

print("Binnen-tier vs tussen-tier op overlappende uurreeksen")
display(ts_pair_summary_df.round(4))

print("Afstand versus similarity")
display(distance_corr_summary.round(6))

print("Pair-level distance/similarity detail")
display(distance_similarity_df.sort_values(["dataset_label", "dist_km"]).round(4))
```

**Interpretatie**

Binnen-tier correlaties liggen gemiddeld boven tussen-tier correlaties, maar het verschil is beperkt. Dat suggereert dat tier nuttig is als grove segmentatie, maar onvoldoende als volledige representatie van parkingdynamiek.

De afstand-similarity relatie is in shortterm negatief (dichterbij = vaker gelijkaardiger profiel), maar in longterm zwak en statistisch onzeker door het kleine aantal parkings. Dit ondersteunt een voorzichtige, niet-deterministische interpretatie van ruimtelijke nabijheid.

**Hypothese status**
- H-S3: **Partially supported** (richting klopt, effect niet dominant).
- H-S7: **Partially supported / inconclusive** (alleen shortterm toont duidelijk afstandssignaal).

**Volgende logische stap**

## 6. Eventeffect per tier en eventtype

```python
EVENT_COLS = [
    "is_event_day",
    "is_football_day",
    "is_sport_day",
    "is_festival_day",
    "is_procession_day",
    "is_kermis_day",
    "is_markt_day",
    "is_carnival_day",
    "is_other_day",
]


event_effect_rows = []
for dataset_label, df in filtered_dfs.items():
    for event_col in EVENT_COLS:
        if event_col not in df.columns:
            continue

        for tier in ["centrum", "vesten_of_rand"]:
            sub = df.loc[df["tier_temporal"] == tier].copy()
            ev = as_flag(sub[event_col])

            n_event = int(ev.sum())
            n_non_event = int((~ev).sum())
            if n_event < 30 or n_non_event < 30:
                continue

            mean_event = float(sub.loc[ev, "occupancy_rate"].mean())
            mean_non_event = float(sub.loc[~ev, "occupancy_rate"].mean())

            event_effect_rows.append(
                {
                    "dataset_label": dataset_label,
                    "event_col": event_col,
                    "tier_temporal": tier,
                    "n_event": n_event,
                    "n_non_event": n_non_event,
                    "mean_event": mean_event,
                    "mean_non_event": mean_non_event,
                    "delta_event_minus_non": mean_event - mean_non_event,
                }
            )

event_effect_df = pd.DataFrame(event_effect_rows)

event_pivot_rows = []
for dataset_label, grp in event_effect_df.groupby("dataset_label"):
    pivot = grp.pivot(index="event_col", columns="tier_temporal", values="delta_event_minus_non")
    pivot["tier_gap_centrum_minus_outer"] = pivot.get("centrum", np.nan) - pivot.get("vesten_of_rand", np.nan)
    pivot["dataset_label"] = dataset_label
    event_pivot_rows.append(pivot.reset_index())

event_pivot_df = pd.concat(event_pivot_rows, ignore_index=True)

print("Eventeffecten per tier")
display(event_effect_df.sort_values(["dataset_label", "event_col", "tier_temporal"]).round(4))

print("Tier-verschil in eventeffect")
display(event_pivot_df.sort_values(["dataset_label", "tier_gap_centrum_minus_outer"], ascending=[True, False]).round(4))
```

## 7. Centrumsaturatie en uitwijkgedrag

```python
def saturation_effect_table(df: pd.DataFrame, thresholds=(0.85, 0.90, 0.95), lags=(0, 1, 2, 3, 6)) -> pd.DataFrame:
    ts = (
        df.groupby(["rounded_hour", "tier_temporal"])["occupancy_rate"]
        .mean()
        .unstack("tier_temporal")
    )
    ts = ts[["centrum", "vesten_of_rand"]].dropna()

    rows = []
    for thr in thresholds:
        sat_flag = ts["centrum"] >= thr
        for lag in lags:
            outer_series = ts["vesten_of_rand"].shift(-lag)
            valid_sat = sat_flag & outer_series.notna()
            valid_non = (~sat_flag) & outer_series.notna()

            n_sat = int(valid_sat.sum())
            n_non = int(valid_non.sum())
            if n_sat < 20 or n_non < 20:
                continue

            rows.append(
                {
                    "threshold": thr,
                    "lag_hours": lag,
                    "n_sat": n_sat,
                    "n_non_sat": n_non,
                    "outer_mean_when_sat": float(outer_series[valid_sat].mean()),
                    "outer_mean_when_not_sat": float(outer_series[valid_non].mean()),
                    "delta_outer_sat_minus_not": float(outer_series[valid_sat].mean() - outer_series[valid_non].mean()),
                }
            )

    return pd.DataFrame(rows)


def share_based_saturation_table(df: pd.DataFrame, thresholds=(0.85, 0.90, 0.95), min_share=0.50, lags=(0, 1, 2, 3)) -> pd.DataFrame:
    center = df.loc[df["tier_temporal"] == "centrum"].copy()
    outer = df.loc[df["tier_temporal"] == "vesten_of_rand"].copy()

    rows = []
    for thr in thresholds:
        center_share = (
            center.assign(sat=center["occupancy_rate"] >= thr)
            .groupby("rounded_hour")["sat"]
            .mean()
            .rename("share_center_sat")
        )
        outer_mean = (
            outer.groupby("rounded_hour")["occupancy_rate"]
            .mean()
            .rename("outer_mean")
        )
        ts = pd.concat([center_share, outer_mean], axis=1).dropna()

        sat_flag = ts["share_center_sat"] >= min_share

        for lag in lags:
            outer_shift = ts["outer_mean"].shift(-lag)
            valid_sat = sat_flag & outer_shift.notna()
            valid_non = (~sat_flag) & outer_shift.notna()

            n_sat = int(valid_sat.sum())
            n_non = int(valid_non.sum())
            if n_sat < 20 or n_non < 20:
                continue

            rows.append(
                {
                    "threshold": thr,
                    "min_share_center_sat": min_share,
                    "lag_hours": lag,
                    "n_sat": n_sat,
                    "n_non_sat": n_non,
                    "delta_outer_sat_minus_not": float(outer_shift[valid_sat].mean() - outer_shift[valid_non].mean()),
                }
            )

    return pd.DataFrame(rows)


def lead_lag_corr_table(df: pd.DataFrame, lags=range(-6, 7)) -> pd.DataFrame:
    ts = (
        df.groupby(["rounded_hour", "tier_temporal"])["occupancy_rate"]
        .mean()
        .unstack("tier_temporal")
    )[["centrum", "vesten_of_rand"]].dropna()

    rows = []
    for lag in lags:
        if lag < 0:
            corr_val = ts["centrum"].shift(-lag).corr(ts["vesten_of_rand"])
        else:
            corr_val = ts["centrum"].corr(ts["vesten_of_rand"].shift(-lag))

        rows.append({"lag_hours": int(lag), "corr_centrum_vs_outer": float(corr_val)})

    return pd.DataFrame(rows)


saturation_rows = []
share_sat_rows = []
lead_lag_rows = []

for dataset_label, df in filtered_dfs.items():
    sat_tbl = saturation_effect_table(df)
    sat_tbl["dataset_label"] = dataset_label
    saturation_rows.append(sat_tbl)

    share_tbl = share_based_saturation_table(df, min_share=0.50)
    share_tbl["dataset_label"] = dataset_label
    share_sat_rows.append(share_tbl)

    ll_tbl = lead_lag_corr_table(df)
    ll_tbl["dataset_label"] = dataset_label
    lead_lag_rows.append(ll_tbl)

saturation_df = pd.concat(saturation_rows, ignore_index=True)
share_saturation_df = pd.concat(share_sat_rows, ignore_index=True)
lead_lag_df = pd.concat(lead_lag_rows, ignore_index=True)

print("Saturatie op basis van centrumgemiddelde")
display(saturation_df.round(4))

print("Saturatie op basis van aandeel centrumparkings >= threshold")
display(share_saturation_df.round(4))

print("Lead-lag correlaties centrum vs vesten_of_rand")
display(lead_lag_df.round(4))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5), sharey=True)

for ax, dataset_label in zip(axes, ["shortterm", "longterm"]):
    sub = saturation_df.loc[saturation_df["dataset_label"] == dataset_label].copy()
    if sub.empty:
        ax.set_title(f"{dataset_label}: onvoldoende saturatiecases")
        continue

    sns.lineplot(
        data=sub,
        x="lag_hours",
        y="delta_outer_sat_minus_not",
        hue="threshold",
        marker="o",
        ax=ax,
    )
    ax.axhline(0, color="black", linewidth=1)
    ax.set_title(f"Delta outer occupancy bij centrumsaturatie ({dataset_label})")
    ax.set_ylabel("delta_outer_sat_minus_not")

plt.tight_layout()
plt.show()

fig, axes = plt.subplots(1, 2, figsize=(16, 4), sharey=True)
for ax, dataset_label in zip(axes, ["shortterm", "longterm"]):
    sub = lead_lag_df.loc[lead_lag_df["dataset_label"] == dataset_label]
    sns.barplot(data=sub, x="lag_hours", y="corr_centrum_vs_outer", color="#4c72b0", ax=ax)
    ax.set_title(f"Lead-lag corr centrum vs outer ({dataset_label})")
    ax.axhline(0, color="black", linewidth=1)

plt.tight_layout()
plt.show()
```

**Interpretatie**

In shortterm stijgt de bezetting in `vesten_of_rand` duidelijk wanneer centrum op hoge saturatiedrempels zit, zowel gelijktijdig als in korte lags. In longterm zijn mean-based saturatiecases schaars, maar de share-based check wijst in dezelfde richting met kleinere maar consistente deltas.

Omdat dit observaties op geaggregeerd niveau zijn, kunnen we dit enkel interpreteren als descriptieve co-beweging, niet als bewijs van causale uitwijk.

**Hypothese status**
- H-S5: **Partially supported** (sterk descriptief signaal in shortterm; in longterm indicatief maar dunner bewijs door beperkte saturatie-incidentie).

**Volgende logische stap**
- Hypotheses samenvatten en vertalen naar modelling-/feature-engineeringimplicaties.

### Reflectie op spatial heterogeniteit voor Phase 3

De spatial heterogeniteit is sterk genoeg om feature engineering te rechtvaardigen, maar vooral in combinatie met parking-specifieke context. `Tier` is nuttig als coarse feature, maar niet voldoende als enige spatial representatie.

Concreet:
1. Gebruik `parking_id`-historische baselines (gemiddelde, q95, volatiliteit) naast `tier` en `parking_location_category`.
2. Neem capaciteit op, maar modelleer interacties met parking/tier in plaats van een uniforme lineaire capaciteitseffectaanname.
3. Gebruik eventfeatures met tier- of parkinginteracties, omdat effecten heterogeen en contextafhankelijk zijn.
4. Beschouw saturatie- en pressure-features (bijv. centrummean, aandeel centrum > 90%) als potentiële spilloverproxies.
5. Vermijd causale spatial claims; interpreteer deze features als associatieve signalen.

## Key findings
1. Centrum versus outer geeft een duidelijk niveauverschil in bezetting, maar binnen-tier heterogeniteit blijft groot.
2. Capaciteit verklaart bezetting niet uniform over datasets; relationele richting is datasetafhankelijk.
3. Parkingprofielen clusteren slechts gedeeltelijk per tier: parking-specifieke modellering lijkt noodzakelijk.
4. Event- en saturatiepatronen tonen relevante spatial interacties, maar met beperkte causaliteitszekerheid.

## Implications for feature engineering
1. Combineer coarse spatial features (`tier`, `location_category`) met parking-level statistische profielen.
2. Voeg interactiefeatures toe: `event × tier`, `event × parking`, `time × tier`.
3. Ontwikkel saturatie-/pressure-indicatoren op systeemniveau (centrumdruk) als kandidaatspilloverfeatures.
4. Gebruik dataset-specifieke feature treatment voor shortterm en longterm waar spatial relaties uiteenlopen.
